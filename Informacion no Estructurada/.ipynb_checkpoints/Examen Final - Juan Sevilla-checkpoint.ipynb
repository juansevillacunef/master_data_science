## 6.2 Utilice un modelo de transformers para realizar un resumen del texto e imprima el resultado


```python
from transformers import EncoderDecoderModel, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("IIC/marimari-r2r-mlsum")
model = EncoderDecoderModel.from_pretrained("IIC/marimari-r2r-mlsum")

input_ids = tokenizer(text[:512], return_tensors="pt").input_ids
output_ids = model.generate(input_ids)[0]

print("Resumen:")
print(tokenizer.decode(output_ids, skip_special_tokens=True))
```

# Pregunta 8
El fichero bbc-text.csv contiene varias noticias ya categorizados. Para responder esta pregunta, realice un notebook en python que realice las siguientes tareas:

Lea el contenido de bbc-text.csv  sobre un DataFrame
Obtenga la matriz Tf-Idf

A partir de la matriz Tf-Idf obtenida en el paso anterior, realice una clasificación utilizando los algoritmo Naive Bayesiano y SVM y obtenga la matriz de confusión correspondiente, y las medidas de precisión del entrenamiento. Adjunte el notebook obtenido.

Nota: Ajuste los valores de los parámetros para generar la matriz Tf-idf como considere más adecuado. Se trata de un clasificación en múltiples categorías. Considere si es conveniente eliminar stop-words, cuál es el número de features que le parece más conveniente, etc.

```python
bbc = pd.read_csv("/Users/juan/Documents/Juan's MacBook Pro/CUNEF/Quinto/Informacion no Estructurada/bbc-text-1.csv")
bbc.head()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.model_selection import train_test_split

modellist=['Bayesiano Ingenuo', 'SVC'] 

def run_models(texto_tfidf, label, modellist, disp_labels = None):
    """ 
        Este funcion entrena cada modelo definido en lista input, separa entre train y test,
        predice las clases para el conjunto test y calcula el accuracy de entrenamiento y test,
        asi como la matriz de confusion par cada modelo.
        
        Input: 
            texto_tfidf: matriz de TFIDF del Corpus
            label: columna con el label de la clase de cada documento. Tienen que ser numéricas
            modellist: lista de strings con los modelos a entrenar
            disp_labels: labels de las clases para las matrices de confusion
            
        Output:
            matriz de confusion, accuacy de entrenaiento, test y fiabilidad para cada modelo entrenado
    """
    plt.rcParams["figure.figsize"] = (20,3)
    
    texto_train, texto_test, label_train, label_test = train_test_split(texto_tfidf, label, train_size = 0.8,\
                                                                        random_state=100, stratify=label)
    
    for name in modellist:
        if name == "Bayesiano Ingenuo":
            model = MultinomialNB()
        if name == "SVC":
            model = SVC(kernel='linear')


        model.fit(texto_train, label_train)
        categs_pred = model.predict(texto_test)
        acc_test = model.score(texto_test, label_test)
        

        cm = confusion_matrix(label_test, categs_pred)
        TP = cm[0][0]
        TN = cm[1][1]
        FP = cm[0][1]
        FN = cm[1][0]

        acc_test = (TP+TN)/(TP+TN+FP+FN)
        pre_test = TP/(TP+FP)
        rec_test = TP/(TP+FN)
        f1_test = (2*pre_test*rec_test)/(pre_test+rec_test)
        
        ConfusionMatrixDisplay.from_predictions(label_test, categs_pred, display_labels = disp_labels);
        

        # Confusion Matrix
        print('Para el modelo','\033[1m' + str(name) + '\033[0m',"matriz de confusion es:")
        plt.show()
        print("Con un accuracy de test de:", round(acc_test*100,2),"%")
        print("Con una precison de test de:", round(pre_test*100,2),"%")
        print("Con un recall de test de:", round(rec_test*100,2),"%")
```

```python
# Corremos ambos modelos sin limite de features y con los stop words
from sklearn.feature_extraction.text import TfidfVectorizer
bbc_tfidf = TfidfVectorizer().fit_transform(bbc["text"])
run_models(bbc_tfidf, bbc["category"], modellist)
```

```python
# Corremos ambos modelos con limite de features de 20 y con los stop words
bbc_tfidf20 = TfidfVectorizer(max_features=20).fit_transform(bbc["text"])
run_models(bbc_tfidf20, bbc["category"], modellist)
```

```python
# Corremos ambos modelos con limite de features de 20 y con los stop words
stopwords = nlp_es.Defaults.stop_words
bbc["text_without_stopwords"] = list(bbc["text"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)])))

bbc_tfidfSW = TfidfVectorizer().fit_transform(bbc["text_without_stopwords"])
run_models(bbc_tfidfSW, bbc["category"], modellist)

```
