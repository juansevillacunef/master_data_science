# Análisis de dependencias

```python
### Importación de las librerías

import spacy
from spacy import displacy
import pandas as pd


nlp = spacy.load("es_core_news_lg")
```

```python
### Función de análisis de dependencias

def get_dependency_parse(doc):
    df=pd.DataFrame(columns=["Token", "POS", "Función", "Descripción"])
    
    for token in doc:
        fila = {'Token':token.text, 'POS':token.pos_, 'Función':token.dep_, 'Descripción':spacy.explain(token.dep_)}
        df = df.append(fila, ignore_index=True)
    return(df)
```

```python
### Principal

frases = 'El conductor del coche rojo sufrió un síncope y se chocó con varios coches aparcados'
doc = nlp(frases)

df=get_dependency_parse(doc)
df
```

```python
# Diagrama de Dependencias

for sent in doc.sents:
    displacy.render(sent, jupyter=True, options={'distance': 110,'arrow_stroke': 2,'arrow_width': 8})
```
