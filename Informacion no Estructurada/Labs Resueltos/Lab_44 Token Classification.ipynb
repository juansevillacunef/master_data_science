# Token Classification (POS y NER)

```python
## Importación de librerías

import spacy
import pandas as pd
from transformers import pipeline

nlp_parsing = spacy.load('es_core_news_lg')
```

```python
## Lectura de fichero

texto_original = open("/Users/juan/Documents/Juan's MacBook Pro/CUNEF/Quinto/Informacion no Estructurada/esp.txt", "r").read()
print(texto_original)
```

```python
## POS _ spaCy

texto_spacy = nlp_parsing(texto_original)

spacy_pos_tagged = [(token.text, token.pos_) for token in texto_spacy]
pd.DataFrame(spacy_pos_tagged, columns=['Token', 'POS']).T
```

```python
## POS _ Transformers

from transformers import AutoTokenizer, AutoModelForTokenClassification

tokenizer = AutoTokenizer.from_pretrained("sagorsarker/codeswitch-spaeng-pos-lince")
model = AutoModelForTokenClassification.from_pretrained("sagorsarker/codeswitch-spaeng-pos-lince")

nlp_pos = pipeline("ner", model = model, tokenizer = tokenizer)

texto_trf = nlp_pos(texto_original)
trans_pos_tagged = [(token["word"], token["entity"]) for token in texto_trf]
pd.DataFrame(trans_pos_tagged, columns=['Token', 'POS']).T
```

```python
## NER _ spaCy

ents_tagged = [(ent.text, ent.label_) for ent in texto_spacy.ents]
pd.DataFrame(ents_tagged, columns=['Entidad', 'Tipo']).T
```

```python
## ENT _ Transformers

from transformers import AutoTokenizer, AutoModelForTokenClassification

tokenizer = AutoTokenizer.from_pretrained("sagorsarker/codeswitch-spaeng-ner-lince")
model = AutoModelForTokenClassification.from_pretrained("sagorsarker/codeswitch-spaeng-ner-lince")

nlp_pos = pipeline("ner", model = model, tokenizer = tokenizer)

texto_trf = nlp_pos(texto_original)
trans_pos_tagged = [(token["word"], token["entity"]) for token in texto_trf]
pd.DataFrame(trans_pos_tagged, columns=['Token', 'POS']).T
```
