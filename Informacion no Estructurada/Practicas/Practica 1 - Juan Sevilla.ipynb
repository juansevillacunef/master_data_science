### Importamos librerias

```python
### Importación de librerías

import spacy
from spacy import displacy
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

nlp = spacy.load("es_core_news_lg")
```

```python
## Leemos el CSV
texto_hotel = pd.read_csv("/Users/juan/Documents/Juan's MacBook Pro/CUNEF/Quinto/Informacion no Estructurada/Practica 1/hotel.csv",) 
```

```python
## Importamos los stopwords del paquete spacy

stopwords = nlp.Defaults.stop_words
print("Hay", str(len(stopwords)), "stopwords")
```

```python
## Pasamos todo el texto a minuscula 

texto_hotel["text"] = texto_hotel["text"].str.lower()
```

Esto se hace para que no haya tokens duplicados (misma palabra con mayuscula y minuscula) en nuestros analisis

```python
## Duplicamos la columna de texto pero sin stopwords

texto_hotel['text_without_stopwords'] = texto_hotel["text"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
texto_hotel
```

Tener esta columna adicional nos permite ver cuant informacion nos estan aportando los stopwords. 
Hay dos posibilidades:
   1. que los stopwords solo sean ruido blanco en cuy caso habria que eliminarlos.
   2. que los stopwords aporten informacion sobre lo que se escribe en la reseña.
   


```python
## Creamos el BOW y lo presentamos como una matriz con cada token como columna y fila la frcuencia en cada doc
vectorizer = CountVectorizer(max_features=20)
BOW = vectorizer.fit_transform(texto_hotel['text'])

vocab = vectorizer.get_feature_names_out()
BOWdf = pd.DataFrame(BOW.toarray(), columns=vocab)
BOWdf
# .sum().sort_values(ascending = False)[:20]
```

```python
## De la misma forma que el BOW creamos un dataframe legible con el resultado del TFIDF

tfifd_vec = TfidfVectorizer(max_features=20)
TFIDF = tfifd_vec.fit_transform(texto_hotel["text"])

vocab = tfifd_vec.get_feature_names_out()
TFIDFdf = pd.DataFrame(TFIDF.toarray(), columns=vocab)
TFIDFdf
# .sum().sort_values(ascending = False)[:20]
```

Como era de esperar, tanto en el BOW como en el TFIDF vemos que de los 20 tokens con mayor importancia en las reseñas, casi todos son stopwrods, sin embargo no parece que sean ruido blacnco. Si tomamos por ejemplo los 10 docmentos que aparecen por defecto en cada matriz (los 5 primeros y los 5 últimos) podemos obtener informacion al contrastar el resultado de cada token en el BOW y TFIDF. <br> <br>
Por ejemplo, el stopword que mas me ha llamado la atencion es la palabra "la", en donde vemos una gran frequencia en las reseñas 1 y 4, pero mucho menor en el resto. Sin embargo, cuando miramos la misma columna en el TFIDF, no sigue el mismo patron. Vemos que en el documento 4 solo hay un 0.38 de importancia, mientras que en el documento 199 con una frecuqncia de 5 hay una importancia de 0.77. <br><br>
Vamos a ver las reseñas 4 y 199 para observar que pueden significar los resultados:

### Viendo que la mayoria de las palabras mas frecuentes son stopwords, vamos a realizar el mismo analisis pero con la columna sin stopwords

```python
BOW = vectorizer.fit_transform(texto_hotel['text_without_stopwords'])

vocab = vectorizer.get_feature_names_out()
pd.DataFrame(BOW.toarray(), columns=vocab)
```

```python
TFIDF = tfifd_vec.fit_transform(texto_hotel['text_without_stopwords'])

vocab = tfifd_vec.get_feature_names_out()
pd.DataFrame(TFIDF.toarray(), columns=vocab)
```

En este caso ya se puede observar que las palabras con mas importancia segun el BOW y TFIDF son palabras que nos dicen algo sobre las reviews que estamos analizando. Podemos observar que elementos de la experiencia del hotel son los que mas destacan los cliente. Además, todos los adjetivos entre las palabras as importantes son postivos (agradable, amable, excelente, etc.) <br> <br>
En **conclusion**, el analisis con stopwords nos permite ver que patron de palabras siguen las reseñas, ya que hemos visto que un determinante como "la" con gran importancia puede señalar una enumeración, mientras que el analisis sin stopwords nos da informacion sobre el propio contenido (si la reseña es positiva o negativa por ejemplo)
