````python
import PyPDF2
import re

# open the pdf file
pdfFileObj = open("/Users/juan/Documents/Juan MacBook Pro/CUNEF/Quinto/TFM/Mates tratado/2019-2020 Modelos examen Matem ticas II.pdf", 'rb')

# create a pdf reader object
pdfReader = PyPDF2.PdfReader(pdfFileObj)

# get the number of pages in the pdf
num_pages = len(pdfReader.pages)

# define the regex pattern
pattern = r'[A-Z]\.\d+\..'
#r'[A-Z]\d+[A-Z]\d+'

# loop over all the pages
Qs = []
for page_num in range(num_pages):
    pageObj = pdfReader.pages[page_num] 
    text = pageObj.extract_text()
    matches = re.findall(pattern, text)
    for match in matches:
        print(match)
        Qs.append(match)

````

````python
pattern_questions = r"[A-Z]\.\d+\.\sCalificación"
pattern_answers = r'[A-Z]\.\d+'
#r'[A-Z]\d+[A-Z]\d+'

# loop over all the pages
Qs = []
QAs = {}
matches = re.findall(pattern_questions, raw["content"])
for match in matches:
    Qs.append(match)

for i in range(0,len(Qs)):
    start_pattern = Qs[i]
    if start_pattern == Qs[-1]:
        end_pattern = "SOLUCIONES"
    else:
        end_pattern = Qs[i+1]

    start_index = raw["content"].find(start_pattern)
    end_index = raw["content"].find(end_pattern)

    text_between_patterns = raw["content"][start_index:end_index]
    QAs[str(start_pattern)[:4]] = {"question": text_between_patterns}

    
As = []
matches = re.findall(pattern_answers, raw["content"])
for match in matches:
    As.append(match)
    
for i in range(int(len(As)/2),len(As)):
    start_pattern = As[i]
    start_index = raw["content"].find(start_pattern) + len(start_pattern)
    text_after_patterns = raw["content"][start_index:]
    start_index = text_after_patterns.find(start_pattern)
    
    if start_pattern == As[-1]:
        text_between_patterns = text_after_patterns[start_index:]
        
    else:
        end_pattern = As[i+1]
        end_index = text_after_patterns.find(end_pattern)
        text_between_patterns = text_after_patterns[start_index:end_index]
        
    if text_between_patterns[5:17] == "Calificación": 
        continue
      
    else:
        QAs[str(start_pattern)[:4]] = {"answer": text_between_patterns}


print(QAs)

#     start_index = raw["content"].find(start_pattern) + len(start_pattern)
#     end_index = raw["content"].find(end_pattern)

#     text_between_patterns = raw["content"][start_index:end_index]

#     print(text_between_patterns)
````
